```python
import pandas as pd
dataset = pd.read_csv("CKD.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
independent=dataset[['age','bp','al','su','bgr','bu','sc','sod','pot','hrmo','pc_normal','pcc_present','ba_present','htn_yes','dm_yes','cad_yes','appet_yes','pe_yes','ane_yes']]
dependent=dataset[['classification_yes']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
param_grid = {
    'criterion': ['gini', 'entropy', 'log_loss'],        # impurity measures
    'max_depth': [None, 5, 10, 20, 30],                  # depth of the tree
    'min_samples_split': [2, 5, 10],                     # min samples to split a node
    'min_samples_leaf': [1, 2, 4],                       # min samples at leaf
    'max_features': [None, 'sqrt', 'log2'],              # features to consider at split
    'ccp_alpha': [0.0, 0.01, 0.05]                        # post-pruning parameter (cost complexity)
}

grid=GridSearchCV(DecisionTreeClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,y_train)
```

```python
re=grid.cv_results_
grid_predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,grid_predictions)
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(y_test,grid_predictions)
print(clf_report)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
```
